```python
# Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

```python
# import iris dataset from seaborn

iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
#check uniques species
iris['species'].unique()
```

```python
# drop rows with class label 'versicolor' 
#iris = iris[iris['species'] != 'versicolor']
```

```python
#split the data into features and target

X =iris.drop('species', axis=1)
y=iris['species']
```

```python
y_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
y_changed = y.map(y_map)

# Plot the dataset
plt.scatter(X['sepal_length'], X['petal_length'], c=y_changed, cmap='viridis')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Iris Dataset')
plt.show()
```

```python
# Create a DecisionTreeClassifier model
dt = DecisionTreeClassifier()

# Fit the model to the data
dt.fit(X, y)

# model accuracy
print('Model Accuracy:', dt.score(X, y))
```

```python
# visualize the decision tree

from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))
plot_tree(dt, feature_names= ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], class_names=['setosa', 'versicolor', 'virginica'], filled=True)
plt.show()
```
